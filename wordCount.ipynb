```python
import requests
```

```python
url="https://ia800302.us.archive.org/20/items/romeojuliet01shak/romeojuliet01shak_djvu.txt"
```

```python
resp=requests.get(url)
```

```python
resp
```

```python
data=resp.text
```

```python
type(data)
```

```python
data.split('\n')
```

```python
nospaces=[]
```

```python
for eachline in data.split('\n'):
    if eachline !='':
        nospaces.append(eachline.lower())
        
        
    
```

```python
nospaces
```

```python
wordcount={}
```

```python
for line in nospaces:
    for word in line.split():
        if word not in wordcount:
            wordcount[word]=1
        else:
            wordcount[word]=wordcount[word]+1
            
```
